```python
import numpy as np
import pandas as pd
data = pd.read_csv('OnlineRetail.csv', header=0, encoding='ISO-8859-1')
display(data)
```

```python
#tìm số hàng và số cột của sữ liệu
print('So hang la:', data.shape[0], 'va so cot la:', data.shape[1])
```

```python
# Việt hóa tiêu đề
data.columns = ['So hoa don', 'Ma hang hoa', 'Mo ta', 'So luong', "Ngay", 'Don gia', 'Ma KH', 'Nuoc']
display(data)
```

```python
# Thêm cột tiền vào sau cuối
# object > float > int > NaN
data['Tien'] = np.NaN
display(data)
```

```python
# Thêm cột tiền vào sau cuối

data.insert(6, 'Chiet khau (%)', 3)
data
```

```python
 # thêm 1 sản phẩm hàng 541905
data.loc[541904.8] = ['324345', '34534B', "Noi chien khong dau", 1000, '23/6/2022 20:30', 50.2, 3, 312313.0, 'Vietnam', np.NaN]
data = data.sort_index().reset_index(drop=True)
display(data)
data.tail(20)
```

```python
# tính tiền: Tien = so luong x don gia x chiet khau 
data['Tien'] = data['So luong'] * data['Don gia'] * (100- data['Chiet khau (%)'])/100
data
```

```python
# lưu dữ liệu
data.to_csv('Ketquahomnay.csv')
print('Xuat file thanh cong')
```
