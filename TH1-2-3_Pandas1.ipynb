```python
import numpy as np
import pandas as pd
data = pd.read_csv('FoodPrice_in_Turkey.csv', header=0, encoding='ISO-8859-1')
display(data)
```

```python
# đổi tên 1 cột xác định
data.rename(columns={'Place' : 'Địa điểm'}, inplace=True)

# đối tên nhiều cột
data.rename(columns={'Place' : 'Địa điểm', 'ProductId' : 'Mã SP'}, inplace=True)

# đối tên toàn bộ cột
data.columns = ["Địa điểm", "Mã SP", "Tên SP", "Đơn vị", "Tên đơn vị", "Tháng", "Năm", "Giá"]
data
```

```python
data['cột mới'] = np.nan
data
```

```python
data.insert(8,'Chiết khấu (%)', np.random.randint(0,20,size=(data.shape[0],1)))
display(data)
data.info()
```

```python
data.insert(5,'Số lượng', 10)
data
```

```python
# thêm dòng vào cuối bảng
data.loc[data.shape[0]] = ['Sài Gòn', 3000, "Dưa hấu", 5, 'KG', 10, 6, 2022, 50.000, 0, np.NaN]
display(data)
```

```python
# thêm dòng vào đầu bảng
data.loc[-1] = ['Hà Nội', 3000, "Dưa hấu", 5, 'KG', 10, 6, 2022, 50.000, 0, np.NaN]
# Update index của bảng 
data.index = data.index + 1
# sắp xếp lại bảng
data = data.sort_index()
display(data)
```

```python
# chèn thêm dòng ở vị trí mong muồn phải lấy số nhỏ hơn vị trí tương ứng
# nếu lấy bằng vị trí mong muốn sẽ thành thay thế
data.loc[2.5] = ['Huế', 3000, "Dưa hấu", 5, 'KG', 10, 6, 2022, 50.000, 0, np.NaN]
# Sắp xếp lại các hàng theo đúng số thứ tự từ nhỏ tới lớn
data = data.sort_index().reset_index(drop=True)
display(data)
```

```python
# xóa cột sử dụng del
data2 = data
display(data2)
# xóa cột ngay trên file gốc rất nguy hiểm
#del data2['cột mới']
display(data2)
display(data)
```

```python
# xóa cột sử dụng pop
data3 = data
display(data3)
# xóa cột ngay trên file gốc rất nguy hiểm => hạn chế dùng
#data3.pop('cột mới')
display(data3)
display(data)
```

```python
# xóa sử dụng hàm drop
# df.drop('Tên cột cần xóa', axis=1, inplace=True/False)
data4 = data
data5 = data4.drop('cột mới', axis=1)
display(data)
display(data4)
display(data5)
```

```python
data6 = data.drop(['cột mới', 'Số lượng'], axis=1)
display(data6)
```

```python
# xóa dòng cuối cùng
data6 = data.drop(data.shape[0]-1, axis=0)
display(data6)
```
